```
%%sql

/* We re-create the table to keep */
DROP TABLE IF EXISTS TestScores;

/* Create the Test Scores table. */
CREATE TABLE TestScores (id INTEGER PRIMARY KEY, name VARCHAR(100), score INTEGER);

INSERT INTO TestScores(name, score) VALUES ('Daisy', 92);
INSERT INTO TestScores(name, score) VALUES ('Neil', 60);
INSERT INTO TestScores(name, score) VALUES ('Carol', 85);
INSERT INTO TestScores(name, score) VALUES ('Jack', 92);
INSERT INTO TestScores(name, score) VALUES ('Jessica', 88);
INSERT INTO TestScores(name, score) VALUES ('Ben', 85);
INSERT INTO TestScores(name, score) VALUES ('Khaled', 73);
INSERT INTO TestScores(name, score) VALUES ('Susan', 67);
INSERT INTO TestScores(name, score) VALUES ('Ali', 65);
```

```
%%sql

SELECT * FROM TestScores
```

```
# We can also embed SQL into our python code. 
# In this example, we will load the table data into a dataframe.

result = %sql SELECT * from TestScores
df = result.DataFrame()
df
```

```
# We can also use this data for various visualizations.

import matplotlib.pyplot as plt
plt.figure(figsize=(18,7))

chart = %sql SELECT name, score FROM TestScores
chart.bar();
```
